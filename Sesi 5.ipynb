## Combining Multiple Datasets

```python
#menggabungkan city_data dengan df lain
further_city_data = pd.DataFrame(
{"revenue": [7000, 3400], "employee_count": [2, 2]}, index=["New York", "Barcelona"])
further_city_data
```

```python
city_data
```

```python
all_city = pd.concat([further_city_data, city_data])
all_city
```

```python
city_countries = pd.DataFrame({ "country": ["Holland", "Japan", "Holland", "Canada", "Spain"], "capital": [1, 1, 0, 0, 0]},
index=["Bandung", "Jakarta", "Pekanbaru", "Toronto", "Barcelona"])
city_countries
```

```python
cities = pd.concat([all_city, city_countries], axis = 1)
cities
```

```python
#menggabungkan kota hanya yang muncul di kedua objek df, maka diatur paratemeternya join menjadi inner join
pd.concat([all_city, city_countries], axis = 1, join = 'inner')
```

```python
countries = pd.DataFrame({ "population_millions": [17, 127, 37],
"continent":["Europe", "Asia", "North America"]
}, index=["Bandung", "Jakarta", "Canada"])
countries
```

```python
# joining 2 or more dataframes
pd.merge(cities, countries, left_on='country', right_index = True, how='outer')
```
